```python
from LogicTree import *
#https://lco.global/spacebook/astrobiology/what-life/
```

```python
#The goal is this:
#I write a few definitions and premises, and then the LogicTree 
#automatically generates the arguments that follow from the 
#definitions and premises, and then it prints it out.
#Definition(term, defining_attributes)
```

```python
definition1 = Definition("Dog", ["is an animal", "has 4 legs"])
# So any entity with attribute ("animal with 4 legs") satisfies 
#the definition of a dog. This is equivalent to the statement P->Q 
#where P = X has attribute (animal with 4 legs), and Q = X is a dog.
#Statement(statement_string). The program should take the statement 
#"A cat is an animal with 4 legs" and interpret it correctly as cat 
#has attributes (animal with 4 legs)
```

```python
premise1 = Statement("A cat is an animal and has 4 legs")
```

```python
definitions = [definition1]
premises = [premise1]

myTree = LogicTree(definitions, premises)
```

```python
myTree.grow()# Apply tautologies to premises and definitions to obtain new statements in the tree. For example, the syllogism ((P->Q)&P)->Q where P = X has attribute (animal with 4 legs), and Q = X is a dog. It should see that the substatement (P->Q) is true by the definition, and P is the premise for a cat, so it can conclude Q for the cat. In other words, it should conclude that "A cat is a dog".
```

```python
myTree.save("treefile.csv")
```

```python
#The real test:
definition1 = Definition("living thing", ["is highly organized", "has the ability to acquire materials and energy", "responds to its environment", "reproduces", "adapts"])
definition2 = Definition("Abortion", "Willfully terminating a human fetus or embryo")
definition3 = Definition("To kill", "To willfully cause a living thing to stop living")
definition4 = Definition("A living human", ["is a living thing", "has a complete human genome in each of its cells"])
definition5 = Definition("Innocent", ["has not committed any crimes"])
premise1 = Statement("A human fetus or embryo is highly organized, has the ability to acquire materials and energy, responds to its environment, reproduces, and adapts.")
premise2 = Statement("A human fetus or embryo has not committed any crimes")
premise3 = Statement("A human fetus or embryo has a complete human genome in each of its cells")
```

```python
definitions = [definition1, definition2, definition3, definition4, definition5]
premises = [premise1, premise2, premise3]
abortionTree = LogicTree(definitions, premises)
```

```python
abortionTree.draw_tree()
```

```python
new_statement = Statement("A human fetus or embryo is a living thing.", parents=["D1","P1"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("A human fetus or embryo is a living thing and has a complete human genome in each of its cells.", parents=["S1","P3"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("A human fetus or embryo is a living human.", parents=["D4","S2"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("It should be illegal to kill innocent humans.")
abortionTree.append(new_statement, kind="P")
```

```python
new_statement = Statement("A human fetus or embryo is innocent.", parents=["D5","P2"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("A human fetus or embryo is an innocent living human.", parents=["S3","S4"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Definition("Terminate", "to cause to stop living.")
abortionTree.append(new_statement, kind="D")
```

```python
new_statement = Statement("Abortion willfully causes a human fetus or embryo to stop living.", parents=["D2","D6"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("Abortion kills a human fetus or embryo.", parents=["D3","S6"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("Abortion kills an innocent living human.", parents=["S5","S7"])
abortionTree.append(new_statement, kind="S")
```

```python
new_statement = Statement("Abortion should be illegal.", parents=["P4","S8"])
abortionTree.append(new_statement, kind="S")
abortionTree.print(verbose=True)
abortionTree.draw_tree()
```

```python
abortionTree.draw_tree(detailed=True)
```
